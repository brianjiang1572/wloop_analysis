```python
import xml.etree.ElementTree as ET 
import numpy as np
import matplotlib.pyplot as plt

import errors
import extract_observables
from plotting_envir import Plotter
```

# Different starting points

```python
data_path = "/home/lqcd/brian137/chroma_Wloops/gpu_test/wflow_to_1.5/data/"
wloop_size = np.array(range(1, 14))
smoothing = [0, 0.5, 1, 1.5]
data = []
for i in wloop_size:
    obj = extract_observables.Data_Processing(data_path, start_time = [i,i,i,i])
    obj.file1 = 3550
    obj.off_axis = True
    data.append(obj)

for i in data: i.read_all_files()
values = np.array([i.find_potential_errors() for i in data])
r = data[0].r
```

```python
values.shape
```

# Fixed smoothing, interate over wloop size

```python
nsmooth = 0
plotter = Plotter(f"Smoothing = {smoothing[nsmooth]}", "r [fm]", "V(r)"); ax = plotter.ax

for i in range(0, values.shape[0], 4):
    ax.errorbar(r*0.09, values[i][:,0][nsmooth], yerr = values[i][:,1][nsmooth], fmt = "o", label = f"tsize = {wloop_size[i]}")
plotter.show_legend()
```
