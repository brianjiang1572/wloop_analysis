```python
import xml.etree.ElementTree as ET 
import numpy as np
import matplotlib.pyplot as plt

import errors
import extract_observables
from plotting_envir import Plotter
```

# wflow to 1.5

```python
data_path = "/home/lqcd/brian137/chroma_Wloops/gpu/data/wflow_to_0.5/data"
smoothing = [0, 0.5, 1, 1.5]
data = extract_observables.Data_Processing(data_path, start_time = [7,10,12,13])
data.file1 = 3550
data.off_axis = True
data.read_all_files()
```

```python
output = data.find_potential_errors()
plotter = Plotter(f"Wilson Flow", "r [fm]", "V(r)"); ax = plotter.ax

for i in range(0,4):
    ax.errorbar(data.r*0.09, output[i][0], yerr = output[i][1], fmt = "o", label = f"t = {smoothing[i]}")

plotter.show_legend()
```

```python
output = data.find_force_errors()
plotter = Plotter(f"Wilson Flow", "r [fm]", "F[r]"); ax = plotter.ax

for i in range(0,4):
    dr = data.r[1:] - data.r[:-1]
    ax.errorbar(data.r[:-1]*0.09, output[i][0]/dr, yerr = output[i][1], fmt = "o", label = f"t = {smoothing[i]}")

plotter.show_legend()
```
